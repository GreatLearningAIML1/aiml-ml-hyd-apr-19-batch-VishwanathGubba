```python
#Importing a random sample of data

import pandas as pd
import random
p = 0.01
df = pd.read_csv(
         'blogtext.csv',
         header=0, 
         skiprows=lambda i: i>0 and random.random() > p
)
```

```python
#Text processing

from gensim.parsing.preprocessing import remove_stopwords,strip_multiple_whitespaces,strip_non_alphanum,strip_numeric
def clean_text(text):
    text=text.lower()
    text=strip_non_alphanum(text)
    text=strip_numeric(text)
    text=remove_stopwords(text)
    text=strip_multiple_whitespaces(text)
    return text
```

```python
df['cleaned_text']=df['text'].apply(lambda x:clean_text(x))
```

```python
df.head()
```

```python
df['labels']='['+df['gender'].astype(str)+', '+df['age'].astype(str)+', '+df['topic'].astype(str)+', '+df['sign'].astype(str)+']'
```

```python
df2 = df[['cleaned_text','labels']]
```

```python
df2
```

```python
X = df2.loc[:, df2.columns != 'labels']
Y = df2['labels']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)
```

```python
X_train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(ngram_range=(1, 2))
X_train_dtm = vect.fit_transform(X_train['cleaned_text'])
X_test_dtm = vect.transform(X_test['cleaned_text'])
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
mlb.fit(y_train)
y_train = mlb.transform(y_train)
y_test = mlb.transform(y_test)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver='lbfgs')
clf = OneVsRestClassifier(clf)
```

```python
clf.fit(X_train_dtm, y_train)
```

```python
y_train_pred = clf.predict(X_train_dtm)
y_test_pred = clf.predict(X_test_dtm)
```

```python
from sklearn.metrics import multilabel_confusion_matrix

multilabel_confusion_matrix(y_test,y_test_pred)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

print(accuracy_score(y_test,y_test_pred))
print(recall_score(y_test,y_test_pred,average='macro'))
print(precision_score(y_test,y_test_pred,average='macro'))
print(f1_score(y_test,y_test_pred,average = 'macro'))
```
