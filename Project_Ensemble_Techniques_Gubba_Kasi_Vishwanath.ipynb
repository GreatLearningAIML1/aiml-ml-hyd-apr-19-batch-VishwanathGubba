```python
#Calculating VIF to assess collinearity in the data
#Using VIF of 5 as cut off
#Will run one iteration after the other by removing he variable with highest VIF

#VIF Factor
def VIF(formula,data):
    from patsy import dmatrices
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    y , X = dmatrices(formula,data = data,return_type="dataframe")
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) 
        for i in range(X.shape[1])]
    return(vif.round(1))

vif=VIF("""status ~ mdvp_fo_hz + mdvp_fhi_hz + mdvp_flo_hz + mdvp_jitter + mdvp_jitter_abs + mdvp_rap + mdvp_ppq + jitter_ddp + mdvp_shimmer + mdvp_shimmer_db + shimmer_apq3 + shimmer_apq5 + mdvp_apq + shimmer_dda + nhr + hnr + rpde + dfa + spread1 + spread2 + d2 + ppe""",df)
vif.sort_values('VIF',ascending=False)  
```

```python
#Running again by removing shimmer_apq3
#Running again by removing jitter_ddp
#Running again by removing mdvp_shimmer
#Running again by removing mdvp_jitter
#Running again by removing mdvp_rap
#Running again by removing mdvp_shimmer_db
#Running again by removing shimmer_apq5
#Running again by removing mdvp_ppq
#Running again by removing ppe
#Running again by removing mdvp_jitter_abs
#Running again by removing shimmer_dda
#Running again by removing hnr

#We have good dataset remaining after removing all the above columns


#VIF Factor
def VIF(formula,data):
    from patsy import dmatrices
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    y , X = dmatrices(formula,data = data,return_type="dataframe")
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) 
        for i in range(X.shape[1])]
    return(vif.round(1))

vif=VIF("""status ~ mdvp_fo_hz + mdvp_fhi_hz + mdvp_flo_hz + mdvp_apq + nhr + rpde + dfa + spread1 + spread2 + d2""",df)
vif.sort_values('VIF',ascending=False) 
```

```python
df = df.drop(['shimmer_apq3','jitter_ddp','mdvp_shimmer','mdvp_jitter','mdvp_rap','mdvp_shimmer_db'
              ,'shimmer_apq5','mdvp_ppq','ppe','mdvp_jitter_abs','shimmer_dda','hnr'], axis=1)
```

```python
#Creating X and y for Train Test separation
```

```python
X = df.loc[:, df.columns != 'status']
```

```python
y = df['status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# Standardizing all features

sc=StandardScaler()
sc.fit(X_train.values)

scaled_features = sc.transform(X_train.values)

Sclaed_X_train = pd.DataFrame(scaled_features, index=X_train.index, columns=X_train.columns)
```

```python
scaled_features1 = sc.transform(X_test.values)

Sclaed_X_test = pd.DataFrame(scaled_features1, index=X_test.index, columns=X_test.columns)
```

```python
dt_model = DecisionTreeClassifier(criterion = 'entropy')
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_predict = dt_model.predict(X_test)
```

```python
print(metrics.accuracy_score(y_test, y_predict))
print(metrics.recall_score(y_test, y_predict))
```

```python
confusion_matrix(y_test,y_predict)
```

```python
#Creating a grid for tuning depth and min samples
```

```python
max_depth = np.arange(1,50,1)
min_samples_leaf = np.arange(1,10,1)
dt_model = DecisionTreeClassifier(criterion = 'entropy')
parameters = {'max_depth' : max_depth,'min_samples_leaf':min_samples_leaf}

GS = GridSearchCV(dt_model, parameters, scoring = 'balanced_accuracy', cv=5)
```

```python
GS.fit(X_train, y_train)
```

```python
GS.best_params_
```

```python
GS.best_score_
```

```python
y_predict_gs=GS.predict(X_test)
```

```python
print(metrics.accuracy_score(y_test, y_predict_gs))
print(metrics.recall_score(y_test, y_predict_gs))

#Accuracy and recall on test dataset has increased after regularization
```

```python
confusion_matrix(y_test,y_predict_gs)
```

```python
#Decision trees have tendency to overfit. Decision trees will try to fit till each data point as different leaf in the tree 
#to give best possible result on the training data set. The problem with this apprach is it won't be generalized enough to give 
#good output on test data set or real world data

#To solve this problem we need to prune the tree with the help of various tuning parameters like the depth of the tree or 
#minimum samples in a node for splitting. These parameters will restrict the tree and generalize it 
#so there is no over fitting on unseen data
```
