```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math
```

```python
titan = pd.read_csv('titan.csv', sep=",", header=0)
titan.head()
```

```python
titan['Old Scheme (in thousands)'] = titan['Old Scheme (in thousands)']*1000
titan['New Scheme (in thousands)'] = titan['New Scheme (in thousands)']*1000

#Multiplying with 1000 since the data given is in 1000's
```

```python
titan.head()
```

```python
titan.loc[:,"Old Scheme (in thousands)"].mean()

#Mean of old scheme is 68033.33
```

```python
titan.loc[:,"New Scheme (in thousands)"].mean()

#Mean of new scheme is 72033.33
```

```python
#Alternate Hypothesis: New Scheme is better than Old Scheme
#Null Hypothesis: New Scheme is same as Old Scheme
```

```python
t_statistic, p_value = stats.ttest_rel(titan['Old Scheme (in thousands)'],titan['New Scheme (in thousands)'])
print(p_value/2)

#Divided by 2 because we have to perform right tailed test
```

```python
#p value is greater than 0.05, hence Null hypothesis holds true. There is no significant difference between old and new schemes
```

```python
#New Alternate hypothesis: (Mean of new scheme - Mean of old scheme) > 5000
#New Null Hypothesis: (Mean of new scheme - Mean of old scheme) <= 5000
```

```python
#Type 1 error is the probability of null hypothesis being rejected when it's true. 
#It is always 5% since we are looking at 95% confidence
```

```python
#Calculate t statistic for 5000 difference

#Calculate Standard Error

SE = stats.sem(titan["New Scheme (in thousands)"]-titan["Old Scheme (in thousands)"])

print(SE)
```

```python
#t = [ (x1 - x2) - d ] / SE
t = ((72033.33-68033.33)-5000)/SE
t
```

```python
#Calculating Degrees of Freedom
DF = len(titan)-1
DF
```

```python
#P value by looking in T tables for t value 

0.3501
```

```python
#Power of a test is probability of not rejecting Null Hypothesis when Null Hypothesis is false
#Calculating t crictical value
#Null hypothesis will be rejected when t value is greater than t critical
```

```python
from scipy.stats import t
t_crit = t.ppf(0.95, 29)
print(t_crit)
```

```python
#Mean Diff for t crictical 1.699

MeanDiff = t_crit*SE
MeanDiff
```

```python
#t value for MeanDiff

t2 = (MeanDiff-5000)/SE
t2
```

```python
#P value for t2

0.4038

#Power of the test is 0.4038
```
